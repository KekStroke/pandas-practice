<h2>Working with Dates and Time Series Data</h2>

```python
import pandas as pd
from datetime import datetime
```

```python
# reading data with time series
df = pd.read_csv('data/ETH-USD.csv')
df
```

```python
# formatting column that contains datetime data to datetime format using special formatter
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

df['Date'].head(5)
```

```python
# getting weekday of date using dt class
df['Weekday'] = df['Date'].dt.day_name()

df.head(4)
```

```python
# filtering datetime
filt = (df['Date'] >= pd.to_datetime('2021-10-01')) & (df['Date'] < pd.to_datetime('2021-11-01'))

df[filt]
```

```python
# creating date parser (from string line to datetime)
d_parser = lambda x: datetime.strptime(x, '%Y-%m-%d')

# initializing datetime columns as reading data and assigning parser to them
# also setting Date as the index column
df = pd.read_csv('data/ETH-USD.csv', parse_dates=['Date'], date_parser=d_parser, index_col='Date')

df
```

```python
df['Weekday'] = df.index.day_name()

df.head(4)
```

```python
# using datetime slicing on DataFrame
df.loc['2021-12':]
```

```python
# getting maximum High value for December 2021
df.loc['2021-12']['High'].max()
```

```python
# resampling 'High' column on months, applying max function
highs = df['High'].resample('M').max()

# plotting Series
highs.plot()
```

```python
# Resampling whole DataFrame on Month and applying different aggregation funciton to specific columns 
df.resample('M').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last', 'Volume': 'sum'})
```

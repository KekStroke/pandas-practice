```python
# To convert dectionary to DataFrame use function
# df = pd.DataFrame(dictionary)
```

```python
# creating a subframe using slicing
# loc - location. Slicing can be used, slicing is inclusive and works with labels
df.loc[5:25:5, 'SOComm':'Accessibility']
# subframe, columns are rows should be specified as to lists
df.loc[[5, 10, 15, 20, 25], ['SOComm', 'NEWOtherComms', 'Age', 'Gender', 'Trans', 'Sexuality', 'Ethnicity', 'Accessibility']]

# the same output, iloc (index location) uses indexing. However, slicing here is exclusive
subdf = df.iloc[5:26:5, 36:44]
```

```python
# FataFrame type
# two-dimensional data structure
type(subdf)
```

```python
# check existing columns
print(subdf.columns)
```

```python
# Series 
# one-dimentional array
series1 = df['SOComm']
series2 = df.loc[1]

print(series1[:5], '\n----------------\n', series2[:5])
```

```python
# get some statistics from Series
series1.value_counts()
```

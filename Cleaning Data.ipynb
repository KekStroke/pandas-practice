<h2>Cleaning Data</h2>
<p>Casting Datatypes and Handling Missing Values<p>

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 48)
```

```python
# setting NA values preliminary
na_vals = ['NA', 'Missing']

df = pd.read_csv('data/survey_results_public.csv', index_col='ResponseId', na_values=na_vals)

# to replace all unwanted values to NaN
# df.replace('NA', np.nan, inplace=True)

df.head(3)
```

```python
# to drop rows with any cells of missing values
df.dropna()
# default values: axis='index', how='any'
```

```python
# drop all rows that have all NA is specific subset
subset=['CompTotal', 'CompFreq']
df.dropna(axis='index', how='all', subset=subset).head(3)
```

```python
# drop collunms that contain any NA vales 
df.dropna(axis='columns', how='any').head(5)
```

```python
# hows a truth table of na values
df.isna().head(7)
```

```python
# set some default value instead of NaN values
df.fillna(value='MISSING').head(3)
```

```python
# setting some defaul values instead of NaN values in specified columns
values = {'YearsCode': 0, 'ConvertedCompYearly': 0}
df.fillna(value=values).head(4)
```

```python
# returns types of columns
df.dtypes[:10]
```

```python
# getting set of unique values
df['YearsCodePro'].unique()
```

```python
df['YearsCodePro'].replace('Less than 1 year', 0, inplace=True)
df['YearsCodePro'].replace('More than 50 years', 51, inplace=True)
df['YearsCode'].replace('Less than 1 year', 0, inplace=True)
df['YearsCode'].replace('More than 50 years', 51, inplace=True)
```

```python
# converting column types to float
df['YearsCodePro'] = df['YearsCodePro'].astype('float')
df['YearsCode'] = df['YearsCode'].astype('float')

df.dtypes[:10]
```

```python
# calculating aggregations of columns
print('Mean(YearsCodePro): ',df['YearsCodePro'].mean())
print('Median(YearsCodePro): ',df['YearsCodePro'].median())
print('Mean(YearsCode): ',df['YearsCode'].mean())
print('Median(YearsCode): ',df['YearsCode'].median())
```

<h2>Filtering</h2>

```python
# in order to filter we should create filt variable first. It should be a Series object that contains True and Flase fields 
filt_py = (df['LanguageHaveWorkedWith'].str.contains('Python', na=False)) & ~(df['ConvertedCompYearly'].isnull()) &\
          (df['Age'] != "Prefer not to say")
filt = ~(df['ConvertedCompYearly'].isnull()) & (df['Age'] != "Prefer not to say")
type(filt)
```

```python
# we can use this object with loc method.
df_python = df.loc[filt_py]
df_python.head(5)
```

```python
# we also can choose some columns in addition to filtered rows
df_python = df.loc[filt_py, ['Age', 'ConvertedCompYearly']]
df_all = df.loc[filt, ['Age', 'ConvertedCompYearly']]
print(df_python.head(10))
```

```python
df_python['Age'].value_counts()
```

```python
# function returns average salary of developer in a given DataFrame by given Age string
def get_average_salary_by_age(df, s):
    lst = list(df.loc[(df['Age'] == s), 'ConvertedCompYearly'])
    return sum(lst)/len(lst)
```

```python
ax_indices = pd.Series(["Under 18 years old", "18-24 years old", "25-34 years old", "35-44 years old", "45-54 years old",
              "55-64 years old", "65 years or older"])
y_python = pd.Series([get_average_salary_by_age(df_python, s) for s in ax_indices])
y_all = pd.Series([get_average_salary_by_age(df_all, s) for s in ax_indices])
```

```python
fig, ax = plt.subplots()

ax.fill_between(ax_indices, y_python, y2=y_all, where=(pd.Series(y_python) > pd.Series(y_all)) ,color='#ACD1AF', 
                interpolate=True, alpha=0.5, label='Python devs earn more')
ax.fill_between(ax_indices, y_python, y2=y_all, where=(pd.Series(y_python) < pd.Series(y_all)) ,color='#F47174', 
                interpolate=True, alpha=0.5, label='Python devs earn less')

ax.plot_date(ax_indices, y_python, linestyle='solid', label='Python Developers')
ax.plot_date(ax_indices, y_all, linestyle='solid', label='All Developers')

fig.autofmt_xdate()

ax.set_ylabel('Salary')
ax.set_title('Average Developers Salary by Age')
ax.legend(loc="lower right")

# fig.savefig('AverageDevSalaryByAge(line).png')

fig.tight_layout()

plt.show()
```

```python
width = 0.3

fig, ax = plt.subplots()

ax.fill_between(ax_indices, y_python, where=(pd.Series(y_python) > pd.Series(y_all)) ,color='#ACD1AF', 
                interpolate=True, alpha=0.2, label='Python devs earn more')
ax.fill_between(ax_indices, y_all, where=(pd.Series(y_python) < pd.Series(y_all)) ,color='#F47174', 
                interpolate=True, alpha=0.2, label='Python devs earn less')

barx = np.arange(len(ax_indices))

ax.bar(barx-width/2, y_python, width=width, label='Python Developers')
ax.bar(barx+width/2, y_all, width=width, label='All Developers')

ax.set_xticks(barx, labels=ax_indices)

fig.autofmt_xdate()

ax.set_ylabel('Salary')
ax.set_title('Average Developers Salary by Age')
ax.legend(loc="upper left")

# fig.savefig('AverageDevSalaryByAge(Bar).png')

fig.tight_layout()

plt.show()
```

<h3>There are several methods that allow to perform some functions on values in DataFrames and Series</h3>
<p>Those are:</p>
<ul>
    <li>apply</li>
    <li>map</li>
    <li>applymap</li>
    <li>replace</li>
</ul>

```python
# apply can be applied no Series to perform a function on every item
df['LanguageHaveWorkedWith'].apply(lambda x: len(x.split(';')) if (isinstance(x, str)) else x)
```

```python
# however, apply is applied on every Series in DataFrame
print(df.apply(len, axis='rows')[:4])
print('--------------------')
print(df.apply(len, axis='columns')[:4])

# this doesn't work!
# df.apply(lambda x: len(x.split(';')) if (isinstance(x, str)) else x)
```

```python
# we can use applymap instead
# counting amount of software used/wonted to use by developers
df.loc[:,'LanguageHaveWorkedWith':'LanguageWantToWorkWith'].applymap(lambda x: len(x.split(';')) \
                                                                           if (isinstance(x, str)) else x)
```

```python
# map is used for substituting each value in a Series with another value
print(df['Gender'].value_counts())
print('---------------------------------------------------------------------')
print(df['Gender'].map({'Man': 'Male', 'Woman': 'Female'}).value_counts())
# it changes vlaues other than specified on NaN
```

```python
# replace another method that is used for substituting each value in a Series with another value
print(df['Gender'].value_counts())
print('---------------------------------------------------------------------')
print(df['Gender'].replace({'Man': 'Male', 'Woman': 'Female'}).value_counts())
# it doesn't change vlaues other than specified
```

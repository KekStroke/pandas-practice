```python
import pandas as pd
```

<h2>Aggregation and Groupping</h2>

```python
pd.set_option('display.max_columns', 48)

df = pd.read_csv('data/survey_results_public.csv')
df.head(3)
```

```python
# to get median sallary
df['ConvertedCompYearly'].median()
```

```python
# get different statistics on all Series that contain numerical information
df.describe()
```

```python
# getting education level of developers
df.value_counts('EdLevel', normalize=True)
```

```python
# grouping data by Country column
ctr_groups = df.groupby(['Country'])
```

```python
# getting results of value_counts function for every group we have
ctr_groups['EdLevel'].value_counts()
```

```python
# getting group and applying value_counts with parameters on it 
ctr_groups.get_group('Kazakhstan').value_counts('EdLevel', normalize=True)
```

```python
# on EdLevel column applying value_counts function and choosing group at the end
ctr_groups['EdLevel'].value_counts().loc['Kazakhstan']
```

```python
# getting median salary of the group
ctr_groups['ConvertedCompYearly'].median().loc['Kazakhstan']
```

```python
# applying several aggregation functions on each group
ctr_groups['ConvertedCompYearly'].agg(['median', 'mean']).head(5)
```

```python
# getting number of developers have worked with python for each group
num_py_ctr = ctr_groups['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python', na=False).sum())
num_py_ctr.head(5)
```

```python
# calculating number of devs in every country
num_ctr = df['Country'].value_counts()

# concatinating two Series objects into DataFrame (they have the same indexes)
python_df = pd.concat([num_ctr, num_py_ctr], axis='columns')

# renaming columns so they would make more sense
python_df = python_df.rename(columns={'Country': 'NumDevs', 'LanguageHaveWorkedWith': 'NumPyDevs'})

python_df.head(5)
```

```python
# adding new PyDevRatio column
python_df['PyDevRatio'] = python_df['NumPyDevs']/python_df['NumDevs']*100

# sorting values by two columns
python_df.sort_values(by=['PyDevRatio', 'NumDevs'], ascending=[False, True], inplace=True)

python_df.head(20)
```

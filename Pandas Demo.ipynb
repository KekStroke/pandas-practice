<h2>Part 2</h2>
<p>Dataframes and series</p>

```python
# To convert dectionary to DataFrame use function
# df = pd.DataFrame(dictionary)
```

```python
# creating a subframe using slicing
# loc - location. Slicing can be used, slicing is inclusive and works with labels
df.loc[5:25:5, 'SOComm':'Accessibility']
# subframe, columns are rows should be specified as to lists
df.loc[[5, 10, 15, 20, 25], ['SOComm', 'NEWOtherComms', 'Age', 'Gender', 'Trans', 'Sexuality', 'Ethnicity', 'Accessibility']]

# the same output, iloc (index location) uses indexing. However, slicing here is exclusive
subdf = df.iloc[5:26:5, 36:44]
```

```python
# FataFrame type
# two-dimensional data structure
type(subdf)
```

```python
# check existing columns
print(subdf.columns)
```

```python
# Series 
# one-dimentional array
series1 = df['SOComm']
series2 = df.loc[1]

print(series1[:5], '\n----------------\n', series2[:5])
```

```python
# get some statistics from Series
series1.value_counts()
```

```python
# Series type
print(type(series1))
print(type(series2))
```

<h2>Indexes</h2>
<p>Index is a unique idenfitier of the row</p>

```python
# setting new index col, however, it erases previous index col
df.set_index('Employment', inplace=True)

# after setting index, you can refer to it using loc method
# index column should not be unique
df.loc["Student, full-time"]

# resetting index column, previous index column becomes the first column in DataFrame
df.reset_index(inplace=True)
df.head(3)
```

```python
# setting ResponseId column as an index column 
df = pd.read_csv("data/survey_results_public.csv", index_col='ResponseId')
df.head(3)
```
